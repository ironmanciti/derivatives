### 5.6.2 BSM Option Valuation

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'serif'
from scipy.integrate import quad      

def dN(x):    # pdf of standard normal random variable x
    return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

def N(d):     # cumulative density function of standard normal random variable x
    # definite integral : integrate func from a to b (possibly infinite interval)
    return quad(lambda x: dN(x), -20, d, limit=50)[0]   

def d1f(St, K, t, T, r, sigma):
    d1 = (np.log(St / K) + (r + 0.5 * sigma**2) * (T - t)) / \
         (sigma * np.sqrt(T - t))
    return d1

# St : stock/index level at time t
# K  : strike price
# t  : valuation date
# T  : date of maturity
# r  : risk-free short rate
# sigma : volatility

def BSM_call_values(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    d2 = d1 - sigma * np.sqrt(T - t)
    call_value = St * N(d1) - np.exp(-r * (T - t)) * K * N(d2)
    return call_value

def BSM_put_values(St, K, t, T, r, sigma):
    put_value = BSM_call_values(St, K, t, T, r, sigma) \
                - St + np.exp(-r * (T - t)) * K
    return put_value

def plot_values(function):
    plt.figure(figsize=(10, 8.3))
    points = 100
    St = 100.0   # index level
    K  = 100.0   # 행사가격
    t  = 0.0     # 평가일 (오늘)
    T  = 1.0     # 만기일 (1년)
    r  = 0.05    # 무위험 이자율
    sigma = 0.2  # 변동성
    plt.subplot(221)
    klist = np.linspace(80, 120, points)
    vlist = [function(St, K, t, T, r, sigma) for K in klist]
    plt.plot(klist, vlist)
    plt.grid(True)
    plt.xlabel('strike $K$')
    plt.ylabel('present value')
    
    plt.subplot(222)
    tlist = np.linspace(0.0001, 1, points)
    vlist = [function(St, K, t, T, r, sigma) for T in tlist]
    plt.plot(tlist, vlist)
    plt.grid(True)
    plt.xlabel('maturity $T$')
    
    plt.subplot(223)
    rlist = np.linspace(0, 0.1, points)
    vlist = [function(St, K, t, T, r, sigma) for r in rlist]
    plt.plot(rlist, vlist)
    plt.grid(True)
    plt.xlabel('short rate $r$')
    plt.ylabel('present value')
    plt.axis('tight')
    
    plt.subplot(224)
    slist = np.linspace(0.01, 0.5, points)
    vlist = [function(St, K, t, T, r, sigma) for sigma in slist]
    plt.plot(slist, vlist)
    plt.grid(True)
    plt.xlabel('volitality $\sigma$')
    plt.tight_layout()
```

```python
plot_values(BSM_call_values)
```

```python
plot_values(BSM_put_values)
```

```python
St = 100.0   # index level
K  = 100.0   # 행사가격
t  = 0.0     # 평가일 (오늘)
T  = 1.0     # 만기일 (1년)
r  = 0.05    # 무위험 이자율
sigma = 0.2  # 변동성
print("ATM (S0 = K) call option ", BSM_call_values(100.0, 100.0, t, T, r, sigma))   
print("    (S0 = K) put  option ", BSM_put_values(100.0, 100.0, t, T, r, sigma))
print("ITM (S0 > K) call option ", BSM_call_values(100.0, 90.0, t, T, r, sigma))   
print("    (S0 < K) put  option ", BSM_put_values(100.0, 110.0, t, T, r, sigma))
print("OTM (S0 < K) call option ", BSM_call_values(100.0, 110.0, t, T, r, sigma))   
print("    (S0 > K) put  option ", BSM_put_values(100.0, 90.0, t, T, r, sigma))
```

### time-to-maturity

maturity 가 가까울 수록 call 과 put 모두 가격 올라감

```python
print("call option, maturity 0.2", BSM_call_values(St, K, t, 0.2, r, sigma))  
print("             maturity 0.5", BSM_call_values(St, K, t, 0.6, r, sigma))   
print("put  option, maturity 0.2", BSM_put_values(St, K, t, 0.2, r, sigma))  
print("             maturity 0.5", BSM_put_values(St, K, t, 0.6, r, sigma))   
```

### short rate

무위험 단기 이자율이 오르면 ITM 으로 만료될 가능성이 커지므로 call option 가격 상승, 반면 put option 은 OTM 으로 만료될 가능성이 증가하므로 가격 하락.

```python
print("call option, short rate 0.02", BSM_call_values(St, K, t, T, 0.02, sigma))  
print("             short rate 0.05", BSM_call_values(St, K, t, T, 0.05, sigma))   
print("put option, short rate 0.02", BSM_put_values(St, K, t, T, 0.02, sigma))  
print("             short rate 0.05", BSM_put_values(St, K, t, T, 0.05, sigma))   
```

### Volatility

변동성이 증가하면 call 과 put 모두 ITM 만료 가능성이 증가하므로 둘 다 가격 상승

```python
print("call option, volatility 0.2", BSM_call_values(St, K, t, T, r, 0.2))  
print("             short rate 0.5", BSM_call_values(St, K, t, T, r, 0.5))  
print("put option,  volatility 0.2", BSM_put_values(St, K, t, T, r, 0.2))  
print("             short rate 0.5", BSM_put_values(St, K, t, T, r, 0.5))   
```

### 5.6.3 BSM Call Greeks

```python
import mpl_toolkits.mplot3d as p3
from BSM_option_valuation import d1f, N, dN

def BSM_delta(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    delta = N(d1)
    return delta

def BSM_gamma(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    gamma = dN(d1) / (St * sigma * np.sqrt(T - t))
    return gamma

def BSM_theta(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    d2 = d1 - sigma * np.sqrt(T - t)
    theta = -(St * dN(d1) * sigma / (2 * np.sqrt(T - t))) \
            + r * K * np.exp(-r * (T - t) * N(d2))
    return theta

def BSM_rho(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    d2 = d1 - sigma * np.sqrt(T - t)
    rho = K * (T - t) * np.exp(-r * (T - t)) * N(d2)
    return rho    

def BSM_vega(St, K, t, T, r, sigma):
    d1 = d1f(St, K, t, T, r, sigma)
    vega =  St * dN(d1) * np.sqrt(T - t)
    return vega

def plot_greeks(function, greek):
    St = 100.0   # index level
    K  = 100.0   # 행사가격
    t  = 0.0     # 평가일 (오늘)
    T  = 1.0     # 만기일 (1년)
    r  = 0.05    # 무위험 이자율
    sigma = 0.2  # 변동성
    
    tlist = np.linspace(0.01, 1, 25)
    klist = np.linspace(80, 120, 25)
    V = np.zeros((len(tlist), len(klist)), dtype=np.float)
    for j in range(len(klist)):
        for i in range(len(tlist)):
            V[i,j] = function(St, klist[j], t, tlist[i], r, sigma)
            
    # 3D Plotting
    x, y = np.meshgrid(klist, tlist)
    fig = plt.figure(figsize=(9, 5))
    plot = p3.Axes3D(fig)
    plot.plot_wireframe(x, y, V)
    plot.set_xlabel('strike $K$')
    plot.set_ylabel('maturity $T$')
    plot.set_zlabel('%s(K, T)' % greek)
```

```python
plot_greeks(BSM_delta, 'delta')
```

```python
plot_greeks(BSM_gamma, 'gamma')
```

```python
plot_greeks(BSM_theta, 'theta')
```

```python
plot_greeks(BSM_rho, 'rho')
```

```python
plot_greeks(BSM_vega, 'vega')
```

### 5.6.4 CRR(COX-ROSS-RUBINSTEIN) Model

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'serif'
from BSM_option_valuation import BSM_call_value

#
# Model Parameters
#
S0 = 100.0  # index level
K = 100.0  # option strike
T = 1.0  # maturity date
r = 0.05  # risk-less short rate
sigma = 0.2  # volatility


# Valuation Function
def CRR_option_value(S0, K, T, r, sigma, otype, M=4):
    ''' Cox-Ross-Rubinstein European option valuation.
    Parameters
    ==========
    S0 : float
        stock/index level at time 0
    K : float
        strike price
    T : float
        date of maturity
    r : float
        constant, risk-less short rate
    sigma : float
        volatility
    otype : string
        either 'call' or 'put'
    M : int
        number of time intervals
    '''
    # Time Parameters
    dt = T / M  # length of time interval
    df = math.exp(-r * dt)  # discount per interval

    # Binomial Parameters
    u = math.exp(sigma * math.sqrt(dt))  # up movement
    d = 1 / u  # down movement
    q = (math.exp(r * dt) - d) / (u - d)  # martingale branch probability

    # Array Initialization for Index Levels
    mu = np.arange(M + 1)
    mu = np.resize(mu, (M + 1, M + 1))
    md = np.transpose(mu)
    mu = u ** (mu - md)
    md = d ** md
    S = S0 * mu * md

    # Inner Values
    if otype == 'call':
        V = np.maximum(S - K, 0)  # inner values for European call option
    else:
        V = np.maximum(K - S, 0)  # inner values for European put option

    z = 0
    for t in range(M - 1, -1, -1):  # backwards iteration
        V[0:M - z, t] = (q * V[0:M - z, t + 1] +
                         (1 - q) * V[1:M - z + 1, t + 1]) * df
        z += 1
    return V[0, 0]


def plot_convergence(mmin, mmax, step_size):
    ''' Plots the CRR option values for increasing number of time
    intervals M against the Black-Scholes-Merton benchmark value.'''
    BSM_benchmark = BSM_call_value(S0, K, 0, T, r, sigma)
    m = range(mmin, mmax, step_size)
    CRR_values = [CRR_option_value(S0, K, T, r, sigma, 'call', M) for M in m]
    plt.figure(figsize=(9, 5))
    plt.plot(m, CRR_values, label='CRR values')
    plt.axhline(BSM_benchmark, color='r', ls='dashed', lw=1.5,
                label='BSM benchmark')
    plt.xlabel('# of binomial steps $M$')
    plt.ylabel('European call option value')
    plt.legend(loc=4)
    plt.xlim(0, mmax)
```

```python
plot_convergence(1, 1000, 20)
```

```python
plot_convergence(1, 1000, 25)
```
